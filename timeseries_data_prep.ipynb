# Imports

```python
import pandas as pd 
import numpy as np
```

# Data Load

```python
HNEI_cycles = pd.read_csv("batteryarchive/metadata.csv")
HNEI_cycles = HNEI_cycles[:15]
print(HNEI_cycles.head(20))
```

```python
X_data = []
y_data = []
for i in range(len(HNEI_cycles)):
    df = pd.read_csv("batteryarchive/" + HNEI_cycles.iloc[i]["cell_id"] + "_timeseries.csv")
    df = df[["Discharge_Energy (Wh)"]]
    df = df[~df.isin([np.nan, np.inf, -np.inf]).any(1)]
    df = df[df["Discharge_Energy (Wh)"]>0]
    df = df["Discharge_Energy (Wh)"]
    print(f'File {i+1} of 15')
    for index, row in enumerate(df):
        if (index+1) != len(df):
            if index%10000 == 0:
                print(f'Row {index} of {len(df)} done')
            temp = []
            if index < 10:
                for i in range(10-index+1):
                    temp.append(0.0)
            for i in reversed(range(10 - len(temp))):
                temp.append(float(format(df.iloc[index-i], '.4f')))
            X_data.append(temp)
            y_data.append(float(format(df.iloc[index+1], '.4f')))
```

```python
data = {
    'X_data' : X_data,
    'y_data' : y_data,
}

timeseries_df = pd.DataFrame(data, columns=['X_data', 'y_data'])
timeseries_df.to_csv("timeseries_df.csv")
```
